>## <font color = 'green'>Questions 3</font> 
>Create a Scatter plot showing the meternal mortal of country's in south Asia 2008 Verses 2017.


```python
import matplotlib.pyplot as plt
# matplotlib inline
# import pandas as pd
# df = pd.read_csv("players.csv")

with open('Data.csv','r') as p:
    rows = csv.DictReader(p)

    tweenty_10 = []
    tweenty_17 = []
    colors = []
    count=0
    for data_t in rows:
        count+=1
        #tweenty_10.append(float(data_t['2010']))
        tweenty_10.append((data_t['2008']))
        tweenty_17.append((data_t['2017']))
        
        if 'AFG' in data_t['CountryCode'] :
            colors.append('Violet')
        elif 'BGD' in data_t['CountryCode'] :
            colors.append('red')
        elif 'BTN' in data_t['CountryCode'] :
            colors.append('magenta')
        elif 'IND' in data_t['CountryCode'] :
            colors.append('green')
        elif 'LKA' in data_t['CountryCode'] :
            colors.append('yellow')
        elif 'MDV' in data_t['CountryCode'] :
            colors.append('orange')
        elif 'NPL' in data_t['CountryCode'] :
            colors.append('cyan')
        elif 'PAK' in data_t['CountryCode'] :
            colors.append('black')
        else: colors.append('blue')
        #if(count==100):
         #   break
    plt.xlabel('tweenty_10')
    plt.ylabel('tweenty_17')
    plt.scatter(tweenty_10,tweenty_17, c=colors)
    plt.rcParams["figure.figsize"] = (20,30)
    plt.show()
```

>## <font color = 'green'>Questions 4</font> 
>Display a Bar chart of maternal Mortal ratio from 2000 to 2017 within two countries of South Asia's Region.


```python
# countries in lower income 
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

with open('Data_row_in_column.csv','r') as p:
    transpose = csv.DictReader(p)
    Bars = []
    Height = []
    for data_t in transpose:
        if data_t['Year'] >'2000' and data_t['Year'] <'2017':
            Bars.append(data_t['Year'])
            Height.append(float(data_t['AFG']))
        if data_t['Year'] >'2000' and data_t['Year'] <'2017':
            Bars.append(data_t['Year'])
            Height.append(float(data_t['BGD']))
           
    ticks = list(range(len(Bars))) # produces [0,1,...,n-1] for n bars
    # print(ticks)
    plt.xticks(ticks, Bars, rotation='vertical')
    plt.bar(ticks, Height,color='blue')
    plt.title("Maternal mortal for Afganistan and Bangladesh")
    plt.ylabel('Year')
    plt.ylabel('Maternal Mortal')
    plt.show()
```

>## <font color = 'green'>Questions 5</font> 
>show a Pie Chart of Maternal mortal Ratio of All countries in South Asia.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

with open('Data_row_in_column.csv','r') as p:
    transpose = csv.DictReader(p)
    all_AFG_Mortal = 0
    all_BGD_Mortal = 0
    all_BTN_Mortal = 0
    all_IND_Mortal = 0
    all_LKA_Mortal = 0
    all_MDV_Mortal = 0
    all_NPL_Mortal = 0
    all_PAK_Mortal = 0

    for data_t in transpose:
        if data_t['Year'] >'2000' and data_t['Year'] <'2017':
            #print(data_t['AFG'],data_t['BTN'])
            all_AFG_Mortal += int(data_t["AFG"])
            all_BGD_Mortal += int(data_t["BGD"])
            all_BTN_Mortal += int(data_t["BTN"])
            all_IND_Mortal += int(data_t["IND"])
            all_LKA_Mortal += int(data_t["LKA"])
            all_MDV_Mortal += int(data_t["MDV"])
            all_NPL_Mortal += int(data_t["NPL"])
            all_PAK_Mortal += int(data_t["PAK"])
    
    plt.pie([all_AFG_Mortal, all_BGD_Mortal,all_BTN_Mortal,all_IND_Mortal,
             all_LKA_Mortal,all_MDV_Mortal,all_NPL_Mortal,all_PAK_Mortal],
            labels=['Afganistan','Bangladesh','Bhutan','India','Srilanka','Maldives','Nepal','Pakistan'],
            autopct='%1.1f%%')
    plt.show()
            
```
