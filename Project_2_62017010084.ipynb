>## <font color = 'magenta'>K-NEAREST NEIGHBORS </font>

```python
f = open('Titanic.csv')
titanic = pd.read_csv(f)
ages_count = 0
sum_age = 0
for i in range(len(titanic)):
    if not pd.isnull(titanic.loc[i]['age']):
        ages_count +=1
        sum_age +=titanic.loc[i]['age']
average_age = sum_age/ages_count

Converted_ages = []
Converted_genders = []
Converted_embarkeds = []

for i in range(len(titanic)):
    if pd.isnull(titanic.loc[i]['age']):
        Converted_ages.append(average_age)
    else: Converted_ages.append(titanic.loc[i]['age'])
    if titanic.loc[i]['gender'] == 'M':
        Converted_genders.append(0)
    else: Converted_genders.append(1)
    if titanic.loc[i]['embarked'] == 'Cherbourg':
        Converted_embarkeds.append(0)
    elif titanic.loc[i]['embarked'] == 'Southampton':
        Converted_embarkeds.append(1)
    else: Converted_embarkeds.append(2)
        
titanic['age'] = Converted_ages
titanic['gender'] = Converted_genders
titanic['embarked'] = Converted_embarkeds

# Create training and test sets

num_items = len(titanic)
percentage = 0.95
num_train = int(num_items*percentage)
num_test = num_items - num_train

titanic_Train = titanic[0:num_train]
titanic_Test = titanic[num_train:]

features = ['gender', 'age', 'class', 'fare', 'embarked']
neighbors = 4
classifier = KNeighborsClassifier(neighbors)
classifier.fit(titanic_Train[features], titanic_Train['survived'])
predictions = classifier.predict(titanic_Test[features])

# Calculate accuracy
num_train_KNN = len(titanic_Train)
num_test_KNN = len(titanic_Test)
correct = 0
for i in range(num_test_KNN):
    print('Predicted:', predictions[i], ' Actual:', titanic_Test.loc[num_train_KNN+i]['survived'])
    if predictions[i] == titanic_Test.loc[num_train_KNN+i]['survived']: correct +=1
print('Accuracy:', correct/num_test_KNN)
```
